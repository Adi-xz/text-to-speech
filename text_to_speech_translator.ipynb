```python
!pip install googletrans==4.0.0-rc1 gTTS > /dev/null
from googletrans import Translator
from gtts import gTTS
from IPython.display import Audio, display
def text_to_speech_translator():
    print("Text-to-Speech Translator\n")
    text = input("Enter text to translate and speak: ")
    print("Supported languages include: English (en), Spanish (es), French (fr), Hindi (hi), German (de), Japanese (ja), Korean (ko), Italian (it), Russian (ru), and Chinese (zh-cn).")
    target_lang = input("Enter target language code (e.g. 'es' for Spanish, 'fr' for French, 'hi' for Hindi): ").lower()
    translator = Translator()
    translation = translator.translate(text, dest=target_lang)
    translated_text = translation.text
    print(f"\n Translated text ({target_lang}): {translated_text}")
    tts = gTTS(translated_text, lang=target_lang)
    audio_file = "translated_speech.mp3"
    tts.save(audio_file)
    print("\n Playing translated speech...")
    display(Audio(audio_file, autoplay=True))
text_to_speech_translator()

```
